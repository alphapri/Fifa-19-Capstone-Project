```python
data['Club'].fillna('No Club', inplace = True)
data['Position'].fillna('ST', inplace = True)
```

* For Club variable, we assume that those player doesn't have any club. so, we impute the missing values with no club.
* For Position variable, we are imputing this with 'ST' because 'ST' position is called Striker.

```python
'''These are categorical variables and will be imputed by mode.'''
data['Preferred Foot'].fillna('Right', inplace = True)
data['International Reputation'].fillna(1, inplace = True)
data['Height'].fillna("5'11", inplace = True)
data['Weight'].fillna('200lbs',inplace = True)
data['Wage'].fillna('€200k', inplace = True)
data['Loaned From'].fillna('None', inplace = True)
data['Jersey Number'].fillna(8, inplace = True)
data['Body Type'].fillna('Normal', inplace = True)
data['Contract Valid Until'].fillna(2019, inplace = True)
data['Work Rate'].fillna('Medium/Medium', inplace = True)
data['Joined'].fillna('Jul 1,2018', inplace = True)
data['Value'].fillna('€60k', inplace = True)
data['Real Face'].fillna('No', inplace = True)
```

# 2.2  Converting Weight, Height,Value ,Wage Columns which are in float       format

```python
#for Values
def value_to_int(Value):
    out = Value.replace('€','')
    if 'M' in out:
        out = float(out.replace('M',''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K',''))*1000
    return float(out)

data['Value'] = data['Value'].apply(lambda x: value_to_int(x))
data['Value'].head()
```

```python
#for wages
data['Wage'] = data['Wage'].apply(lambda x: value_to_int(x))
data['Wage'].head()
```

```python
'''For Height'''
def feet_to_inches(value): 
    tmp = value.split("'")
    return int(tmp[0]) * 12 + int(tmp[1]) #converting feet to inches

data['Height'] = data['Height'].apply(lambda x: feet_to_inches(x))
data['Height'].head()
```

```python
# for Weight
def extract_value_from(value):
    out = value.replace('lbs','')
    return float(out)

data['Weight'] = data['Weight'].apply(lambda x : extract_value_from(x))
data['Weight'].head()
```

# 2.3   Converting skills column which are in the str format


```python
data[['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']].head()
```

```python
def skillConverter(val):
    if type(val) == str:
        s1 = val[0:2]
        s2 = val[-1]
        val = int(s1) + int(s2)
        return val
    
    else:
        return val
```

```python
skill_columns = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']
                      
for col in skill_columns:
    data[col] = data[col].apply(skillConverter)
    
a= data[['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM',
       'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM',
       'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB']].head()    
a
```

# 2.4  GROUPING

```python
def general(data):
    return int(round((data[['HeadingAccuracy', 'Dribbling', 'Curve', 'BallControl']].mean()).mean()))

def mental(data):
    return int(round((data[['Aggression', 'Positioning', 'Vision','Composure']].mean()).mean()))

def mobility(data):
    return int(round((data[['Acceleration', 'SprintSpeed', 'Agility','Reactions']].mean()).mean()))

def power(data):
    return int(round((data[['Balance', 'Jumping', 'Stamina', 'Strength']].mean()).mean()))

def shooting(data):
    return int(round((data[['Finishing', 'Volleys', 'FKAccuracy', 'ShotPower','LongShots', 'Penalties']].mean()).mean()))

def passing(data):
    return int(round((data[['Crossing', 'ShortPassing', 'LongPassing']].mean()).mean()))

def defending(data):
    return int(round((data[['Marking','Interceptions', 'StandingTackle', 'SlidingTackle']].mean()).mean()))

def goalkeeping(data):
    return int(round((data[['GKDiving', 'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']].mean()).mean()))
```

'''Adding these categories to main dataframe'''

```python

data['General'] = data.apply(general, axis = 1)
data['Mental'] = data.apply(mental, axis = 1)
data['Mobility'] = data.apply(mobility, axis = 1)
data['Power'] = data.apply(power, axis = 1)
data['Shooting'] = data.apply(shooting, axis = 1)
data['Passing'] = data.apply(passing, axis = 1)
data['Defending'] = data.apply(defending, axis = 1)
data['Goalkeeping'] = data.apply(goalkeeping, axis = 1)

data.head()
```

# Method 1 : ML Regression of Selected data 

```python
data["Wage"].describe()
```

```python
wage = data[['International Reputation','Age','Overall','Potential','Value','Wage']]
```

```python
# Extracting X and y

X = wage.iloc[:,:-1]
y = wage.iloc[:,-1]
```

# 1.Linear Regression

```python
# Train and test split:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Training the Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Making Predictions
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
# Evaluate Algorithms
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', 
      np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

rmse_LR = round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)), 2) #This variable is created for using it in table of
rmse_LR                                                                 #                            algorithm comparison.
```

Since RMSE is greater than the 10% of mean of wages(i.e.973), it is not a correct metric to evaluate algorithms in our problem.

```python
# Calculate MAE Percentage
mape = 100 * (abs(y_pred - y_test) / y_test)
mape
```

```python
np.mean(mape)
```

```python
regressor.score(X_train, y_train)
```

The R2 score or R-Squared, measures how close the data fits to the regression model, the more the number approaches 1 the more it shows that a significant percentage of the values are explained by the linear regression model - indicating stronger prediction capability

```python
print('R2 score of y_test and y_predict: %.2f' % metrics.r2_score(y_test, y_pred))
r2_LR = round(metrics.r2_score(y_test,y_pred), 2) #This variable is created for using it in table of
                                                  #                            algorithm comparison.
```

It has a good score of 73%
